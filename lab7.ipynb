```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Q1 load the following image and make red rectangle around face and write below rectangle “Cat” with yellow color.

```python
img = cv2.imread('original.png')
font2=cv2.FONT_HERSHEY_COMPLEX
cv2.rectangle(img,(800,130),(1500,800),(255,0,0),3)
#cv2.rectangle(img,(50,500),(1800,1400),(255,255,0),3)
cv2.putText(img,"Cat",(1099,900),font2,4,(255,255,0),10)
plt.imshow(img)
```

# Q2 load the following image and apply sharpen filter.

```python
img = cv2.imread("blurred.png")
kernel = np.array([[-1,-1,-1],
                [-1,9,-1],
                [-1,-1,-1]])
sharpen_img = cv2.filter2D(src=img,ddepth=-1,kernel=kernel)
cv2.imshow('Original', img)
cv2.imshow('Sharpened', sharpen_img)
cv2.waitKey()
```

# Observation: sharpen filter remove more details from original image.

# Q2 load the following image and apply blur filter 

```python
img = cv2.imread("sharpen.png")
#median = cv2.medianBlur(src=img, ksize=5)
gaussian_blur = cv2.GaussianBlur(src=img, ksize=(5,5),sigmaX=0, sigmaY=0)
cv2.imshow("original",img)
cv2.imshow("blurred",gaussian_blur)
cv2.waitKey()
```

# Observation: there's more blurring on original image appered in blurred image but when i apply median filter remove less details and there's less blurring
