```python
import cv2
import datetime
import math
t = datetime.datetime.now()

img = np.zeros((500,500,3),np.uint8)
font2=cv2.FONT_HERSHEY_SIMPLEX
# Date
cv2.putText(img,"Date: ",(7,40),font2,1,(64, 224, 208),2)
cv2.putText(img,t.strftime("%Y-%m-%d"),(100,40),font2,1,(64, 224, 208),2)

# Time
cv2.putText(img,"Time: ",(7,70),font2,1,(64, 224, 208),2)
cv2.putText(img,t.strftime("%H:%M:%S"),(100,70),font2,1,(64, 224, 208),2)

# Circle of Clock
cv2.circle(img,(250,250),140,(0,0,255),2)

# Center of Clock
center = [250,250]

#cv2.circle(img,(250,250),5,(255,0,100),-1)
'''
cv2.line(img, (250,250), (300,320),(255,0,0), 2)
cv2.line(img,(250,250),(300,330),(64, 224, 208),1)
cv2.line(img,(250,250),(300,310),(0,255,0),2)'''
'''angle_hour = (h/12)*360
angle_min = (m/60)*360
angle_sec = (s/60)*360'''

def get_ticks():
    hours = []
    hours_final = []

    for i in range(0,360,6):
        x = int(center[0] + 140 * math.cos(i * math.pi / 180))
        y = int(center[1] + 140 * math.sin(i * math.pi / 180))
        hours.append((x,y))

    for i in range(0,360,6):
        x = int(center[0] + (140-20) * math.cos(i * math.pi / 180))
        y = int(center[1] + (140-20) * math.sin(i * math.pi / 180))
        hours_final.append((x,y))

    return hours, hours_final


#Function for drawing the hands of teh clock
def draw_hand(img):

	#getting the time from time module 
	hour = math.fmod(t.hour, 12)
	minute = t.minute
	second = t.second

	second_angle = math.fmod((second * 6 + 270), 360)
	minute_angle = math.fmod(minute * 6 + 270, 360)
	hour_angle = math.fmod((hour*30) + (minute/2) + 270, 360)

# radius = 140

	second_x = int(center[0] + (140-25) * math.cos(second_angle * math.pi / 180))
	second_y = int(center[1] + (140-25) * math.sin(second_angle * math.pi / 180))
	cv2.line(img, center, (second_x, second_y), (255,255,255), 2)

	minute_x = int(center[0] + (140-60) * math.cos(minute_angle * math.pi / 180))
	minute_y = int(center[1] + (140-60) * math.sin(minute_angle * math.pi / 180))
	cv2.line(img, center, (minute_x, minute_y), (255,0,0), 3)

	hour_x = int(center[0] + (140-100) * math.cos(hour_angle * math.pi / 180))
	hour_y = int(center[1] + (140-100) * math.sin(hour_angle * math.pi / 180))
	cv2.line(img, center, (hour_x, hour_y),(0,0,255), 7)

	cv2.circle(img, center, 5, (255,0,255), -1)
	
	return img


hours, hours_final = get_ticks()

for i in range(len(hours)):
	if i % 5 == 0:
		cv2.line(img, hours[i], hours_final[i], (255,255,255), 3)
		#cv2.putText(img,i,(250,250),font2,1,(255,0,0),2)
		#cv2.putText(img,"Time: ",(7,110),font2,1,(64, 224, 208),2)
	else:
		cv2.circle(img, hours[i], 5, (255,0,100), -1)
# print hours 
cv2.putText(img,"12",(230,100),font2,1,(255,250,250),2) # print 12
cv2.putText(img,"1",(315,115),font2,1,(255,250,250),2) # print 1
cv2.putText(img,"2",(380,180),font2,1,(255,250,250),2) # print 2
cv2.putText(img,"3",(400,260),font2,1,(255,250,250),2) # print 3
cv2.putText(img,"4",(380,340),font2,1,(255,250,250),2) # print 4
cv2.putText(img,"5",(320,405),font2,1,(255,250,250),2) # print 5
cv2.putText(img,"6",(240,425),font2,1,(255,250,250),2) # print 6
cv2.putText(img,"7",(160,405),font2,1,(255,250,250),2) # print 7
cv2.putText(img,"8",(100,340),font2,1,(255,250,250),2) # print 8
cv2.putText(img,"9",(80,260),font2,1,(255,250,250),2) # print 9
cv2.putText(img,"10",(80,180),font2,1,(255,250,250),2) # print 10
cv2.putText(img,"11",(150,115),font2,1,(255,250,250),2) # print 11

#Run
while True:
	image = img.copy()

	#Use draw time to make clock hands on the canvas
	clock = draw_hand(image)

	#Show the watch
	cv2.imshow('clock', image)
	if(cv2.waitKey(1)==ord('q')):
		break

cv2.destroyAllWindows()
cv2.waitKey()
```
