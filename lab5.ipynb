# run video stream using your laptop camera and save frames when i press key “s” close camera when i press “q” ,name of images should be “img.png +{counter}”


```python
import cv2
cap=cv2.VideoCapture(0)
i=0
while cap.isOpened():
    stat, frame=cap.read()
    if stat==True:
        cv2.imshow("my_video",frame)
        k=cv2.waitKey(10)
        print(k)
        if k ==ord('s'):
            cv2.imwrite('img'+str(i)+'.png', frame)
            i+=1
        elif k==ord('q'):
            break
    else:
        cv2.destroyAllWindows()
        print("wrong_frame")
        break
cap.release()
cv2.destroyAllWindows()

```

# Make 100*100 rgb pixel image with black background and make diagonal of image with (x) color such that x is came from rgb trackbar

```python
img=np.zeros((100,100))
#img.shape
#img[:]=np.eye(10,dtype=np.uint8)
def nothing(x):
    print(x)
cv2.namedWindow("my_image")
cv2.createTrackbar('R','my_image',0,255,nothing)    
cv2.createTrackbar('G','my_image',0,255,nothing)    
cv2.createTrackbar('B','my_image',0,255,nothing)
for i,row in enumerate(img):
    for j ,col in enumerate(row):
        if i == j: 
            R = cv2.getTrackbarPos('R','my_image')
            G = cv2.getTrackbarPos('G','my_image')
            B = cv2.getTrackbarPos('B','my_image')
            img[:] = [R,G,B]
            print(i,row)

        cv2.imshow("my_image",img)
        k=cv2.waitKey(1000)
        if k==ord('q'):
            break
print(img)
plt.imshow(img,cmap='gray')
cv2.destroyAllWindows()
```
